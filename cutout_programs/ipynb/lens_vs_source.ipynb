```python
import tensorflow as tf
import numpy as np
from astropy.io import fits
import importlib.util
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
for i in range(20):
    file = Path('challenge_data/GroundBasedTraining/Data/Private/Band1/NoLensImage/imageSDSS_R-10{:04d}.fits'.format(i))
    if file.is_file():
        
        full_band1 = fits.open('challenge_data/GroundBasedTraining/Data/Public/Band1/imageSDSS_R-10{:04d}.fits'.format(i))[0].data
        full_band2 = fits.open('challenge_data/GroundBasedTraining/Data/Public/Band2/imageSDSS_I-10{:04d}.fits'.format(i))[0].data
        full_band3 = fits.open('challenge_data/GroundBasedTraining/Data/Public/Band3/imageSDSS_G-10{:04d}.fits'.format(i))[0].data
        full_band4 = fits.open('challenge_data/GroundBasedTraining/Data/Public/Band4/imageSDSS_U-10{:04d}.fits'.format(i))[0].data
        full = np.stack([full_band1, full_band2, full_band3, full_band4], 2)
        
        nolens_band1 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band1/NoLensImage/imageSDSS_R-10{:04d}.fits'.format(i))[0].data
        nolens_band2 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band2/NoLensImage/imageSDSS_I-10{:04d}.fits'.format(i))[0].data
        nolens_band3 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band3/NoLensImage/imageSDSS_G-10{:04d}.fits'.format(i))[0].data
        nolens_band4 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band4/NoLensImage/imageSDSS_U-10{:04d}.fits'.format(i))[0].data
        nolens = np.stack([nolens_band1, nolens_band2, nolens_band3, nolens_band4], 2)

        nosource_band1 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band1/NoSourceImage/imageSDSS_R-10{:04d}.fits'.format(i))[0].data
        nosource_band2 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band2/NoSourceImage/imageSDSS_I-10{:04d}.fits'.format(i))[0].data
        nosource_band3 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band3/NoSourceImage/imageSDSS_G-10{:04d}.fits'.format(i))[0].data
        nosource_band4 = fits.open('challenge_data/GroundBasedTraining/Data/Private/Band4/NoSourceImage/imageSDSS_U-10{:04d}.fits'.format(i))[0].data
        nosource = np.stack([nosource_band1, nosource_band2, nosource_band3, nosource_band4], 2)

        plt.figure(figsize=(15,10))
        
        plt.subplot(141)
        plt.title('image {} (full)'.format(i))
        plt.imshow(full[:,:,0])
        plt.subplot(142)
        plt.imshow(full[:,:,1])
        plt.subplot(143)
        plt.imshow(full[:,:,2])
        plt.subplot(144)
        plt.imshow(full[:,:,3])
        
        plt.figure(figsize=(15,10))

        plt.subplot(241)
        plt.title('image {} with no lens'.format(i))
        plt.imshow(nolens[:,:,0])
        plt.subplot(242)
        plt.imshow(nolens[:,:,1])
        plt.subplot(243)
        plt.imshow(nolens[:,:,2])
        plt.subplot(244)
        plt.imshow(nolens[:,:,3])

        plt.subplot(245)
        plt.title('image {} with no source'.format(i))
        plt.imshow(nosource[:,:,0])
        plt.subplot(246)
        plt.imshow(nosource[:,:,1])
        plt.subplot(247)
        plt.imshow(nosource[:,:,2])
        plt.subplot(248)
        plt.imshow(nosource[:,:,3])
    else:
        print(file, 'does not exist: moving onto next one...')
```
