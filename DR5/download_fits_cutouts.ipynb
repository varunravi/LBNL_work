```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt
import tensorflow as tf
import importlib.util
from pathlib import Path
import urllib.request
import wget
```

```python
url = 'http://legacysurvey.org/viewer/fits-cutout?ra=190.1086&dec=1.2005&layer=decals-dr5&pixscale=0.27&bands=grz'
filename = wget.download(url)
fitsfile = fits.open(filename)
image = fitsfile[0].data
```

```python
plt.figure(figsize=(10,7))
plt.subplot(131)
plt.imshow(image[0,:,:])
plt.subplot(132)
plt.imshow(image[1,:,:])
plt.subplot(133)
plt.imshow(image[2,:,:])
```

```python
ra = [190.2086, 191.2086]
dec = [1.3005, 1.4005]
for r, d in zip(ra, dec):
    url = 'http://legacysurvey.org/viewer/fits-cutout?ra={}&dec={}&layer=decals-dr5&pixscale=0.27&bands=grz'.format(r, d)
    filename = wget.download(url)
    image = fits.open(filename)[0].data
    
    plt.figure(figsize=(10,7))
    plt.subplot(131)
    plt.imshow(image[0,:,:])
    plt.subplot(132)
    plt.imshow(image[1,:,:])
    plt.subplot(133)
    plt.imshow(image[2,:,:])
    
```
